Itamaracá implimented in python.
Technique by
Henrique Pereira, D. (2022). Itamaracá: A Novel Simple Way to Generate Pseudo-random Numbers. Cambridge Open Engage. doi:10.33774/coe-2022-zsw6t 


```python
import numpy as np
import matplotlib.pyplot as plt
def ita(n, s0, s1, s2, square, iterations):
    """
    Psuedo-random number generator using ablsolute value functions
    Technique by Pereira Haneique
    
    Generates a list of random numbers from 0 < x < 1 
    
    Args:
    n int:
    
        maximum value of function
    s0, s1, s2 int:
        seed numbers need to be smaller then n
    square float:
        float between 1 < x < 4
    iterations int:
        number of random numbers to generate
    """
    #p is part of calculation, FN is random number, sn list is seed number values
    snlist = [s0, s1, s2]
    p = 0 
    FN = 0
    results = []
    for i in range(iterations):
        p = np.absolute((snlist[2]-snlist[1])+(snlist[1]-snlist[0]))
        #needs square root of number between 1<x<4
        FN = int(np.absolute(n-(p*np.sqrt(3.1459))))
        #dividing FN/n gives you a float ranging from 0 < x < 1
        results.append(FN/n)
        #adds FN to newest sn number, then pops oldest number
        snlist.append(FN)
        snlist.pop(0)
    return results

results = ita(10000, 10000*np.random.random(),10000*np.random.random(), 10000*np.random.random(), 3.1459, 2000)
x = range(len(results))
plt.plot(x, results)
plt.show
```
