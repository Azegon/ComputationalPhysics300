```python
import numpy as np
import matplotlib.pyplot as plt

tau = 3.053*60
atoms = 2000
seconds = 1000
mu = np.log(2)/tau
pb = np.zeros([seconds-1])
tl = atoms * np.ones([seconds-1])

for i in range(atoms):
    z0 = np.random.random()
    x0 = int(-np.log(1-z0)/mu) 
    if x0<seconds:
        pb[x0:] += 1
        tl[x0:] -= 1
        
plt.plot(tl, label = 'Tl')
plt.plot(pb, label = 'Pb')
plt.xlabel("Time (s)")
plt.ylabel("Number of atoms")

plt.legend()
plt.show()
```
