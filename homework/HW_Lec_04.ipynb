```python
from numpy import linspace, zeros, absolute
from matplotlib import pyplot as plt


#test parameters
x1,x2,xres = -2, 2, 2000
y1,y2,yres = -2, 2, 2000
n = 100

x_set = linspace(x1,x2,xres)
y_set = linspace(y1,y2,yres)

#grid array
mandset = zeros([xres,yres],float)

#calculations
for i in range(len(y_set)):
    for l in range(len(x_set)):
        #starting conditions
        z=0
        c = x_set[l] + y_set[i]*1j
        
        #iterator
        for k in range(n):
            z = z**2+c
            if absolute(z) >= 2:
                mandset[i,l] = k
                break
                

fig = plt.figure(figsize=(12, 12))
plt.imshow(mandset, origin="lower", extent=[x1,x2,y1,y2], cmap='Greys')
plt.show()
```
