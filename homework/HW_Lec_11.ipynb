```python
import numpy as np
import matplotlib.pyplot as plt
from math import pi
from random import random
import scipy

#using desmos, the interection is 1.6

def MC(n, f):
    """
    This function runs the monte carlo integral of given function
    !only works on HW 11!
    
    Args:
    n: int
        number of iterations
    f: lambda
        mathmatical function for monte carlo integral
    """
    count = 0
    for i in range(n):
        ##change to change range of rectangle
        limx = 1.6
        limy = 1.0
        #random point generation
        x = limx*random()
        y = limy*random()
        if y < f(x):
            count += 1
    i = (limx*count)/n
    return i
    
def runMC(f, g, n, t):
    """
    Runs mc t number of times on given functions and returns list of values
    !only works on HW11!
    """
    values = []
    for i in range(t):
        #final integral is f-g 
        values.append(MC(n, f)- MC(n, g))
        
    return values


#function 1
f = lambda x: np.sqrt(1-(x-1)**2)
#function 2
g = lambda z: 2-np.sqrt(4-z**2)


x = [100,1000,10000,100000]
stdList = []

for i in x:
    values = runMC(f, g, i, 100)
    print('Monte Carlo Integral of', i)
    print('Average of integral:', np.average(values)) 
    print('standard deviation:',np.std(values), '\n')
    stdList.append(np.std(values))

plt.plot(x, stdList)
plt.title('Standard Deviance of Monte Carlo Integrak')
plt.xlabel('Iterations')
plt.ylabel('Standard Deviation')
plt.show
    
```
